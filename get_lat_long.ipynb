```python
# import libraries
import pandas as pd
import requests
import json
```

Read causes of death csv

```python
df = pd.read_csv('data_files/cause_of_deaths.csv')
```

API call to mapquest to get latitude and longitude values

```python
def get_lat_long(df):
    for i, country in enumerate(df['Country/Territory']):
        parameters = {
            'key' : 'NWzfLixXzrP0GAwppmmDDqrTtMf4xVVK',
            'location' : country
        }
        response = requests.get('http://www.mapquestapi.com/geocoding/v1/address', params = parameters)
        data = json.loads(response.text)['results'][0]['locations'][0]['latLng']
            
        lat = data['lat']
        long = data['lng']
        
        df.at[i, 'lat'] = lat
        df.at[i, 'lng'] = long
        
get_lat_long(df)
```

```python
df.to_csv('data_files/cause_of_deaths_geo.csv')
```
